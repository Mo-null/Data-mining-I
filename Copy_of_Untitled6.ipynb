```python

import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
df = pd.read_csv(url)
X = StandardScaler().fit_transform(df.iloc[:, 1:])
```

```python

eps_values = [0.3, 0.5, 0.7, 1.0]
min_samples_values = [3, 5, 7, 10]

print("DBSCAN Results:")
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X)

        non_noise_mask = labels != -1
        non_noise_labels = labels[non_noise_mask]
        non_noise_X = X[non_noise_mask]

        n_clusters = len(set(non_noise_labels))
        noise_points = np.sum(labels == -1)

        unique, counts = np.unique(non_noise_labels, return_counts=True)
        cluster_sizes = dict(zip(unique, counts))

        if n_clusters > 1:
            sil_score = silhouette_score(non_noise_X, non_noise_labels)
        else:
            sil_score = -1

        print(f"eps={eps:.1f}, min_samples={min_samples}: "
              f"clusters={n_clusters}, noise={noise_points}, "
              f"sizes={cluster_sizes}, silhouette={sil_score:.3f}")
```
