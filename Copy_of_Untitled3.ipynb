```python

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

df1 = pd.read_excel("https://raw.githubusercontent.com/Mo-null/Data-mining-I/main/market_basket_reduced.xlsx")
df2 = pd.read_excel("https://raw.githubusercontent.com/Mo-null/Data-mining-I/main/supermarket_sales_reduced.xlsx")

print(" Datasets imported successfully")
```

```python

df1_encoded = df1.notna().astype(int)

if 'Product' in df2.columns and 'Invoice ID' in df2.columns:
    baskets = df2.groupby('Invoice ID')['Product'].apply(list).values.tolist()
    te = TransactionEncoder()
    df2_encoded = pd.DataFrame(te.fit_transform(baskets), columns=te.columns_)
else:
    df2_encoded = df2.notna().astype(int)

print(" Data prepared successfully for Apriori algorithm")
```

```python

def run_apriori(df, support, confidence):
    frequent_itemsets = apriori(df, min_support=support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
    return len(frequent_itemsets), len(rules), rules

print("=== PART A: 50% Support, 75% Confidence ===")
freq1a, rules1a, rules_obj1a = run_apriori(df1_encoded, 0.5, 0.75)
freq2a, rules2a, rules_obj2a = run_apriori(df2_encoded, 0.5, 0.75)

print(f"Dataset 1: {freq1a} frequent itemsets, {rules1a} association rules")
print(f"Dataset 2: {freq2a} frequent itemsets, {rules2a} association rules")
print(" Part A completed successfully")
```

```python

print("=== PART B: 60% Support, 60% Confidence ===")
freq1b, rules1b, rules_obj1b = run_apriori(df1_encoded, 0.6, 0.6)
freq2b, rules2b, rules_obj2b = run_apriori(df2_encoded, 0.6, 0.6)

print(f"Dataset 1: {freq1b} frequent itemsets, {rules1b} association rules")
print(f"Dataset 2: {freq2b} frequent itemsets, {rules2b} association rules")
print(" Part B completed successfully")
```

```python

print("=== FINAL RESULTS ===")
print(f"Dataset 1 - Part A: {rules1a} rules, Part B: {rules1b} rules")
print(f"Dataset 2 - Part A: {rules2a} rules, Part B: {rules2b} rules")

if rules1a > 0 and len(rules_obj1a) > 0:
    sample = rules_obj1a.iloc[0]
    print(f"Sample Rule: {list(sample['antecedents'])} -> {list(sample['consequents'])}")
    print(f"Support: {sample['support']:.3f}, Confidence: {sample['confidence']:.3f}, Lift: {sample['lift']:.3f}")

print(f"Total Rules: {rules1a + rules1b + rules2a + rules2b}")
print(" Apriori completed - support, confidence, lift computed for pattern evaluation")
```
